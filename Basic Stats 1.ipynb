```python
num_data.mean() #Mean calculates the average value of the data
```

```python
num_data.median() #Median gives the middle value in the data
```

```python
num_data.mode().iloc[0] #mode give the unique and most repeated value
```

```python
num_data.std() #It shows how consistent or scattered the data is (Here it is scattered)
```

```python
num_data.var()
```

●	Box Plots:

	Discuss any findings, such as extreme values or unusual distributions.

●	Bar Chart Analysis for Categorical Column:

	Identify categorical columns in the dataset.

```python
num_data.hist(figsize=(30,40),bins=30) 
```

It is showing right skew(Greater than zero,positive) for everything except for discount rate Discount rate has left skew(Less than zero,negative) so in histogram discount rate plot only changed and the outliers of discount rate is also on left

```python
num_data.skew()
```

```python
num_data.boxplot(figsize=(10,10))
```

Here in the boxplot we can see Total Sales value and Net Sales Value has extreme values while the volume and discount rate is consistent

```python
import matplotlib.pyplot as plt
for i in obj_data.columns:
    obj_data[i].value_counts().plot(kind='bar',color='blue',figsize=(6,6))
    plt.show()
```
